```python
from pathlib import Path

from stagpy.stagyydata import StagyyData

sdat = StagyyData(Path("/path/to/run"))

import numpy as np  
import matplotlib.pyplot as plt

n = 3000 #Snapshot number
```

## Temperature Profile:

```python
snapshot = sdat.snaps[n]

T_field = snapshot.fields["T"]            # "T" = name of the temperature field
temp_values = T_field.values               # numpy array with the field data
temp_meta = T_field.meta                   # metadata about the field
            
unit = T_field.meta.dim  # calculate unit

time = sdat.snaps[n].timeinfo["t"] 
time_myr = time / (1e6 * 365.25 * 24 * 3600)


from stagpy import field as sp_field
fig, ax, mesh, cbar = sp_field.plot_scalar(snapshot, "T")

ax.text(
    0, 0,                              # x, y in data coordinates
    f"{time_myr:.2f} Myr",                 # text to draw
    ha="center", va="center",          # horizontal & vertical alignment
    fontsize=18, color="black",        # choose color/font
    bbox=dict(facecolor="white", alpha=0.7, edgecolor="none")  # optional background
)

cbar.set_label(f"Temperature [{unit}]") #unit in colorbar label

fig.set_size_inches(10, 6)
fig.savefig("temperature.png", dpi=400)
```
